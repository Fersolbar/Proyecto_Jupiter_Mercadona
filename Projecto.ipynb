En primer lugar tenemos tres dataframes:
- Raw data nos indica la categoría, nombre, precio y nombre del supermercado
- Compras nos indica el id_producto y el id_cliente
- Competitors nos indica el nombre del supermercado que esta referenciado con la tabla Raw Data.



##Podemos realizar un join entre la columnas, pero debemos nombrar la columna igual

```python
df_sin_nulos = df_sin_nulos.rename(columns={'id_reference_supermarket': 'ID','id_product': 'id_producto'})
```

```python
resultado_1 = pd.merge(df_sin_nulos, df_competitors, on='ID', how='inner')
```

#AHORA TENEMOS NUESTRO DATAFRAME FINAL

```python
resultado_final = pd.merge(resultado_1, df_compras, on='id_producto', how='inner')
resultado_final
```
